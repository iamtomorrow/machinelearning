```python
# Matplotlib Subplots
# Draw 2 plots:
import matplotlib.pyplot as plt
import numpy as np

# 1st plot
x = np.array([1, 3, 5])
y = np.array([3, 5, 2])

plt.subplot(1, 2, 1)
plt.plot(x, y)

# 2nd plot
x = np.array([1, 4, 6])
y = np.array([0, 3, 6])

plt.subplot(1, 2, 2)
plt.plot(x, y)

plt.show()
```

```python
# Draw 2 plots on top of each other:
import matplotlib.pyplot as plt
import numpy as np

# 1st plot:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(2, 1, 1)
plt.plot(x, y)

# 2nd plot:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(2, 1, 2)
plt.plot(x, y)

plt.show()
```

```python
# Draw 6 plots:
import matplotlib.pyplot as plt
import numpy as np

# 1st plot:
x = np.array([1, 2, 3])
y = np.array([3, 4, 0])

plt.subplot(2, 3, 1)
plt.plot(x, y)

# 2nd plot:
x = np.array([1, 2, 3])
y = np.array([2, 7, 4])

plt.subplot(2, 3, 2)
plt.plot(x, y)

# 3th plot:
x = np.array([1, 2, 3])
y = np.array([2, 0, 8])

plt.subplot(2, 3, 3)
plt.plot(x, y)

# 4th plot:
x = np.array([1, 2, 3])
y = np.array([4, 1, 2])

plt.subplot(2, 3, 4)
plt.plot(x, y)

# 5th plot:
x = np.array([1, 2, 3])
y = np.array([8, 6, 1])

plt.subplot(2, 3, 5)
plt.plot(x, y)

# 6th plot:
x = np.array([1, 2, 3])
y = np.array([5, 3, 1])

plt.subplot(2, 3, 6)
plt.plot(x, y)

plt.show()
```

```python
# Title
# 2 plots, with titles:
import matplotlib.pyplot as plt
import numpy as np

# 1st plot:
x = np.array([1, 2, 3, 4])
y = np.array([3, 2, 8, 5])

plt.subplot(1, 2, 1)
plt.plot(x, y)
plt.title('SALES')

# 2nd plot:
x = np.array([1, 2, 3, 4])
y = np.array([40, 20, 10, 30])

plt.subplot(1, 2, 2)
plt.plot(x, y)
plt.title("INCOME")

plt.show()
```
