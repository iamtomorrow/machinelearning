```python
# Matplotlib Line
# Use a dotted line:
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, linestyle = 'dotted')
plt.show()
```

```python
# Use a dashed line:
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, linestyle = 'dashed')
plt.show()
```

```python
# Shorter Syntax
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, ls = ':')
plt.show()
```

```python
# Use a dashdot line:
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, ls = '-.')
plt.show()
```

```python
# Line Color
# Set the line color to red:
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, color = 'red')
plt.show()
```

```python
# Plot with the color named "hotpink":
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, linestyle='-.', c='hotpink')
plt.show()
```

```python
# Line Width
# Plot with a 20.5pt wide line:
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, linewidth='20.5', c='red')
plt.show()
```

```python
# Multiple Lines
# Draw two lines by specifying a plt.plot() function for each line:
import matplotlib.pyplot as plt
import numpy as np

y1 = np.array([3, 8, 1, 10])
y2 = np.array([6, 2, 7, 11])

plt.plot(y1)
plt.plot(y2)

plt.show()
```

```python
# Draw two lines by specifiyng the x- and y-point values for both lines:
import matplotlib.pyplot as plt
import numpy as np

xpoints1 = np.array([1, 3, 5, 8])
ypoints1 = np.array([3, 8, 1, 10])

xpoints2 = np.array([1, 3, 5, 8])
ypoints2 = np.array([6, 2, 7, 11])

plt.plot(xpoints1, ypoints1, xpoints2, ypoints2)

plt.show()
```
